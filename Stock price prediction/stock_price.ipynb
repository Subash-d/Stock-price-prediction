```python
import yfinance as yf
import pandas as pd
```

```python
stock_price  = yf.download('TATAMOTORS.NS', start= '2022-03-01', end= '2025-07-20')
```

```python
df = pd.DataFrame(stock_price )
df.head()
```

```python
df.columns = [col[0] for col in df]
```

```python
vals = ['Close', 'High', 'Low', 'Open', 'Volume']
df[vals] = df[vals].round(2)
```

```python
df = df.drop(columns = ['level_0', 'index'])
```

```python
df.to_csv('stock_price.csv', index = False)
```
