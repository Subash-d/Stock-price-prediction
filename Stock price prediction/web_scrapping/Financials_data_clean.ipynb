```python
import pandas as pd
df = pd.read_csv('Financials.csv')
```

```python
df.head()
```

```python
df = df.drop('Graph', axis=1)  # Unwanted column
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.dropna()
df.drop(index= 18,inplace= True)    # Null row(unwanted)
df.head()
```

```python
df
```

```python
df['Indicator'] = df['Indicator'].str.replace(r'^.*?_Q', '', regex=True)
```

```python
df['Indicator'] = df['Indicator'].str.replace(r'^.*?_TTM','', regex=True)
df['Indicator']
```

```python
df.columns = [col.replace("'", "") for col in df.columns]
```

```python
df.tail()
```

```python
df = pd.concat([df['Indicator'], df.drop('Indicator', axis= 1).iloc[:, ::-1]], axis = 1)
df
```

```python
# Droping unwanted indicators that didn't matters 

final_df = df[~df['Indicator'].isin(['Operating Rev.', 'Operating Exp.', 'Net profit TTM', 'Basic EPS TTM'])].reset_index(drop=True)
```

```python
#  Only indicators that has potential to affect the stock market
final_df
```

```python
final_df.to_csv('Financials_clean.csv', index = 0)
```
